### Task 1: Introduction

#### (a) Engineering problem

The main problem we study here is **irregularities in intracellular traffic in axons** caused by **mutations in molecular motors**. This can lead to **“traffic jams”** (local accumulation of transported cargo). In addition, such jams may cause **axon swelling**, which is associated with various **neurodegenerative diseases**. We treat this as an **engineering problem** and build a **mathematical model** of traffic-jam formation during **molecular-motor-assisted transport** of intracellular organelles, using **transport equations** developed by **Smith and Simmons**.


In an axon, cargo can either diffuse freely in the cytoplasm or bind to molecular motors and move directionally along microtubules (advection). Attachment and detachment between these states are modeled by reaction rates that couple the concentrations. Therefore, transport equations are suitable because they combine diffusion/advection with local gain–loss terms to describe how cargo concentrations evolve in space and time.


#### (b) Mathematical model, variables, and parameters

We model three concentrations (depending on position $x\in[0,L]$ and time $t\ge 0$):

- $n(x,t)$: concentration of **free** cargo/particles in the cytoplasm (diffusing),
- $n_+(x,t)$: concentration of cargo **bound to kinesin**, moving in the **+ direction**,
- $n_-(x,t)$: concentration of cargo **bound to dynein**, moving in the **− direction**.

The coupled transport model is:
$$
n_t = D_0\, n_{xx} - (k_+ + k_-)\,n + k_p\,n_+ + k_n\,n_-, \qquad x\in[0,L],
$$
$$
(n_+)_t + (V_+\, n_+)_x = k_+\,n - k_p\,n_+,
$$
$$
(n_-)_t + (V_-\, n_-)_x = k_-\,n - k_n\,n_-.
$$

**Parameters**
- $D_0$: diffusion coefficient of free cargo,
- $V_+(x,t)$, $V_-(x,t)$: transport velocities along microtubules (for + and − motion),
- $k_+(x,t)$, $k_-(x,t)$: attachment rates from free cargo to the + and − motor-driven states,
- $k_p(x,t)$, $k_n(x,t)$: detachment rates from + and − states back to the free state,
- $L$: axon length (domain size).

#### (c) Purpose of each transport equation and role of the terms

**1) Free cargo equation ($n$):**
$$
n_t = D_0\, n_{xx} - (k_+ + k_-)\,n + k_p\,n_+ + k_n\,n_-.
$$
- $n_t$: how the free concentration changes in time.
- $D_0 n_{xx}$: **diffusion** of free cargo (spreads cargo from high to low concentration).
- $-(k_+ + k_-)n$: **loss** of free cargo due to **attachment** to motors (into $n_+$ or $n_-$).
- $+k_p n_+ + k_n n_-$: **gain** of free cargo due to **detachment** from the motor-driven states.

**2) Plus-moving cargo equation ($n_+$):**
$$
(n_+)_t + (V_+ n_+)_x = k_+ n - k_p n_+.
$$
- $(n_+)_t$: time change of plus-moving cargo.
- $(V_+ n_+)_x$: **advection/transport** of cargo along the axon in the **+ direction** (flux divergence).
- $+k_+ n$: **source** term from attachment of free cargo to the + motor state.
- $-k_p n_+$: **sink** term from detachment back to the free state.

**3) Minus-moving cargo equation ($n_-$):**
$$
(n_-)_t + (V_- n_-)_x = k_- n - k_n n_-.
$$
- $(n_-)_t$: time change of minus-moving cargo.
- $(V_- n_-)_x$: **advection/transport** along the axon in the **− direction**.
- $+k_- n$: **source** term from attachment of free cargo to the − motor state.
- $-k_n n_-$: **sink** term from detachment back to the free state.

Overall, the model combines **diffusion** (spreading of free cargo), **advection** (directed motor transport), and **reaction terms** (attachment/detachment exchange between the three states).

#### (e) Why we need the generic advection equation $u_t + a(x,t)u_x = q(x,t,u)$
The equations for $n_+$ and $n_-$ are **advection–reaction** equations. To connect them to the standard advection form, we expand the flux term:

$$
(V n)_x = V\,n_x + V_x\,n.
$$

For example, the $n_+$ equation becomes:

$$
(n_+)_t + V_+(x,t)\,(n_+)_x
= \underbrace{k_+(x,t)\,n - k_p(x,t)\,n_+ - (V_+)_x(x,t)\,n_+}_{q(x,t,u)}.
$$

This matches the generic form:

$$
u_t + a(x,t)u_x = q(x,t,u),
$$

with $u=n_+$ and $a(x,t)=V_+(x,t)$ (similarly $u=n_-$, $a=V_-$).



### Task 2
### (a) Exact solutions for the linear problems (I), (II), and (III)

We obtain exact solutions using the **method of characteristics**.

---

## Problem (I)
**PDE:**  
$$
u_t + a\,u_x = 0,\qquad a>0,\quad x\in[0,2],\qquad u(x,0)=u_0(x).
$$

### Step 1: Characteristic curves
Along a characteristic $x=x(t)$:
$$
\frac{dx}{dt}=a
\quad\Rightarrow\quad
x(t)=x_0+a t,
\qquad\Rightarrow\qquad
x_0 = x-a t.
$$

### Step 2: Solution along characteristics
$$
\frac{du}{dt} = u_t + \frac{dx}{dt}u_x = u_t + a u_x = 0
\quad\Rightarrow\quad
u(x(t),t)=u(x_0,0)=u_0(x_0).
$$

### Exact solution
$$
\boxed{u(x,t)=u_0(x-a t).}
$$

---

## Problem (II)
**PDE:**  
$$
u_t + x\,u_x = 0,\qquad x\in[0,2],\qquad u(x,0)=u_0(x).
$$

### Step 1: Characteristic curves
$$
\frac{dx}{dt}=x
\quad\Rightarrow\quad
\frac{1}{x}dx = dt
\quad\Rightarrow\quad
\ln|x| = t + C
\quad\Rightarrow\quad
x(t)=x_0 e^{t},
\qquad\Rightarrow\qquad
x_0 = x e^{-t}.
$$

### Step 2: Solution along characteristics
$$
\frac{du}{dt}=u_t + \frac{dx}{dt}u_x = u_t + x u_x = 0
\quad\Rightarrow\quad
u(x(t),t)=u(x_0,0)=u_0(x_0).
$$

### Exact solution
$$
\boxed{u(x,t)=u_0(xe^{-t}).}
$$

---

## Problem (III)
**PDE:**  
$$
u_t + u_x = x,\qquad x\in[0,2],\qquad u(x,0)=u_0(x).
$$

### Step 1: Characteristic curves
$$
\frac{dx}{dt}=1
\quad\Rightarrow\quad
x(t)=x_0+t,
\qquad\Rightarrow\qquad
x_0 = x-t.
$$

### Step 2: ODE for $u$ along characteristics
$$
\frac{du}{dt}=u_t + \frac{dx}{dt}u_x = u_t + u_x = x.
$$
Along the characteristic, $x=x(t)=x_0+t$, so:
$$
\frac{du}{dt}=x_0+t.
$$

### Step 3: Integrate
Integrate from $0$ to $t$:
$$
u(t)-u(0)=\int_0^t (x_0+s)\,ds = x_0 t + \frac{1}{2}t^2.
$$
At $t=0$, $u(0)=u(x_0,0)=u_0(x_0)$, hence:
$$
u(x(t),t)=u_0(x_0)+x_0 t+\frac{1}{2}t^2.
$$

### Exact solution
Substitute $x_0=x-t$:
$$
u(x,t)=u_0(x-t) + (x-t)t + \frac{1}{2}t^2
= u_0(x-t) + x t - \frac{1}{2}t^2.
$$
$$
\boxed{u(x,t)=u_0(x-t)+xt-\frac{1}{2}t^2.}
$$


### (b) Numerical solution and comparison with exact solution (I), (II), (III)

We compute numerical solutions on the domain $x\in[0,2]$ using $M=500$ grid blocks and compare at time $T=0.5$ with the exact solutions derived in part (a).

We use a first-order **upwind finite-volume** method:
- For problems (I) and (III): constant advection speed $a=1$.
- For problem (II): variable speed $a(x)=x$ (still nonnegative on $[0,2]$, so upwind uses the left state).
- For problem (III): there is also a source term $b(x)=x$.

The initial condition is:
$$
u_0(x)=\begin{cases}
1, & x\in[0.8,1.2],\\
0, & \text{otherwise}.
\end{cases}
$$


We choose the time step $\Delta t$ using a CFL condition,
$$
\max_{x\in[0,2]} a(x)\,\frac{\Delta t}{\Delta x} \le \text{CFL},
$$
with CFL $=0.9$. For problem (II), $a(x)=x$ so $\max a = 2$ on $[0,2]$.

Since advection is to the right ($a(x)\ge 0$), we use **upwind (left) states** at cell interfaces and impose an inflow value at $x=0$ (taken as $u(0,t)=0$).

Finally, we compare numerical and exact solutions at $T=0.5$ using plots and an $L^1$ error:
$$
\|e\|_{L^1}\approx \Delta x\sum_j |u_j^{num}-u_j^{exact}|.
$$


```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------
# Grid and initial condition
# -------------------------
M = 500
xL, xR = 0.0, 2.0
T = 0.5

dx = (xR - xL) / M
x_edges = np.linspace(xL, xR, M+1)              # cell interfaces
x = 0.5 * (x_edges[:-1] + x_edges[1:])          # cell centers

def u0_func(xvals):
    """Box function initial data u0(x)=1 on [0.8,1.2], else 0."""
    return np.where((xvals >= 0.8) & (xvals <= 1.2), 1.0, 0.0)

u0 = u0_func(x)

# --------------------------------
# Choose time step (CFL-safe for max speed 2)
# --------------------------------
CFL = 0.9
amax_global = 2.0                      # max speed for a(x)=x on [0,2]
dt_guess = CFL * dx / amax_global
N = int(np.ceil(T / dt_guess))
dt = T / N                             # adjust so we hit exactly T

print(f"dx = {dx:.6f}, dt = {dt:.6f}, N = {N}, CFL_max = {amax_global*dt/dx:.3f}")

# -------------------------
# Upwind solvers
# -------------------------
def solve_I_upwind(u_init, a=1.0):
    """Problem (I): u_t + a u_x = 0 (a>0), upwind scheme."""
    u = u_init.copy()
    for _ in range(N):
        u_in = 0.0                      # inflow boundary at x=0 (not important here at T=0.5)
        u_left = np.empty_like(u)
        u_left[0] = u_in
        u_left[1:] = u[:-1]
        u = u - (dt/dx) * a * (u - u_left)
    return u

def solve_II_upwind(u_init):
    """
    Problem (II): u_t + x u_x = 0  (NON-conservative form)
    Upwind (x>=0): u_j^{n+1} = u_j^n - (dt/dx) * x_j * (u_j^n - u_{j-1}^n)
    """
    u = u_init.copy()
    for _ in range(N):
        u_in = 0.0  # inflow at x=0 (not important here because a(0)=0)
        u_left = np.empty_like(u)
        u_left[0] = u_in
        u_left[1:] = u[:-1]
        u = u - (dt/dx) * (x) * (u - u_left)   # x is the array of cell centers x_j
    return u


def solve_III_upwind(u_init, a=1.0):
    """Problem (III): u_t + a u_x = x, upwind + source term."""
    u = u_init.copy()
    for _ in range(N):
        u_in = 0.0
        u_left = np.empty_like(u)
        u_left[0] = u_in
        u_left[1:] = u[:-1]
        u = u - (dt/dx) * a * (u - u_left) + dt * x   # source b(x)=x
    return u

# -------------------------
# Exact solutions at time T
# -------------------------
def exact_I(xvals, a=1.0):
    return u0_func(xvals - a*T)

def exact_II(xvals):
    return u0_func(xvals * np.exp(-T))

def exact_III(xvals):
    return u0_func(xvals - T) + xvals*T - 0.5*T**2

# -------------------------
# Run all three
# -------------------------
uI_num = solve_I_upwind(u0, a=1.0)
uII_num = solve_II_upwind(u0)
uIII_num = solve_III_upwind(u0, a=1.0)

uI_ex = exact_I(x, a=1.0)
uII_ex = exact_II(x)
uIII_ex = exact_III(x)

# -------------------------
# Errors (optional)
# -------------------------
def L1_error(u_num, u_ex):
    return dx * np.sum(np.abs(u_num - u_ex))

print("L1 errors at T=0.5:")
print("  Problem I : ", L1_error(uI_num, uI_ex))
print("  Problem II: ", L1_error(uII_num, uII_ex))
print("  Problem III:", L1_error(uIII_num, uIII_ex))

# -------------------------
# Plots
# -------------------------
def plot_compare(x, u_num, u_ex, title):
    plt.figure(figsize=(8,4))
    plt.plot(x, u_num, label="numerical")
    plt.plot(x, u_ex, "--", label="exact")
    plt.xlabel("x")
    plt.ylabel("u(x,t)")
    plt.title(title)
    plt.grid(True)
    plt.legend()
    plt.show()

plot_compare(x, uI_num, uI_ex, "Problem (I):  u_t + u_x = 0   at T=0.5")
plot_compare(x, uII_num, uII_ex, "Problem (II): u_t + x u_x = 0 at T=0.5")
plot_compare(x, uIII_num, uIII_ex, "Problem (III): u_t + u_x = x at T=0.5")
```

**Problem (I):** The numerical solution is close to the exact solution at $T=0.5$. The “box” moved to the right, as expected. The edges are not sharp because the first-order upwind method smooths jumps (this is called numerical diffusion).

**Problem (II):** The numerical and exact solutions match well at $T=0.5$. The height stays close to 1, and the box is in the correct place. The edges are again smoother than the exact solution because of numerical diffusion. Small differences near $x=2$ come from the boundary of the domain.

**Problem (III):** The numerical solution is close to the exact one, especially where the curve is smooth. Because of the source term $x$, the solution increases with $x$, and the numerical method captures this. The biggest differences are near the jump edges, which become smooth with the upwind method, and a small difference near the left boundary may come from the chosen boundary value.

## Task 3(a) — From dimensional model (1) to dimensionless model (5)

### Goal
The model is first written in *dimensional* variables $(x,t,n,n_+,n_-)$.  
To simplify the analysis and reduce the number of parameters, we introduce *dimensionless* variables (denoted by hats) and rewrite the PDE system in the dimensionless form (5). Following the lecture notes, we introduce characteristic scales and rewrite the system in dimensionless form.(shortened version).

---

### 1) Dimensional model (starting point)
We start from the dimensional system (model (1)):
$$
\begin{aligned}
n_t &= D_0\,n_{xx}-(k_+ + k_-)n + k_p n_+ + k_n n_-, \\
(n_+)_t + (V^+ n_+)_x &= k_+ n - k_p n_+, \\
(n_-)_t + (V^- n_-)_x &= k_- n - k_n n_-,
\qquad x\in[0,L].
\end{aligned}
$$

---

### 2) Choice of dimensionless variables (scaling)
We introduce the dimensionless variables (given in (6)):
$$
\hat x=\frac{k_+}{V_{+,0}}\,x,\qquad
\hat t=k_+\,t,
$$
$$
\hat n = n\,\frac{V_{+,0}^3}{k_+^3},\qquad
\hat n_\pm = n_\pm\,\frac{V_{+,0}^3}{k_+^3},
$$
and the dimensionless parameters
$$
\hat D_0=\frac{D_0 k_+}{V_{+,0}^2},\qquad
\hat k_-=\frac{k_-}{k_+},\qquad
\hat k_p=\frac{k_p}{k_+},\qquad
\hat k_n=\frac{k_n}{k_+}.
$$
The spatial domain also becomes
$$
x\in[0,L]\quad \Rightarrow \quad \hat x\in[0,\hat L],\qquad
\hat L=\frac{k_+}{V_{+,0}}\,L.
$$

---

### 3) Transforming derivatives
From $\hat t=k_+t$ and $\hat x=\frac{k_+}{V_{+,0}}x$, we get:
$$
\frac{\partial}{\partial t}=k_+\frac{\partial}{\partial \hat t},\qquad
\frac{\partial}{\partial x}=\frac{k_+}{V_{+,0}}\frac{\partial}{\partial \hat x},\qquad
\frac{\partial^2}{\partial x^2}=\left(\frac{k_+}{V_{+,0}}\right)^2\frac{\partial^2}{\partial \hat x^2}.
$$
Also, from the concentration scaling:
$$
n=\hat n\,\frac{k_+^3}{V_{+,0}^3},\qquad
n_\pm=\hat n_\pm\,\frac{k_+^3}{V_{+,0}^3}.
$$

---

### 4) Substitution: first equation
Start from
$$
n_t = D_0\,n_{xx}-(k_+ + k_-)n + k_p n_+ + k_n n_-.
$$
Insert the scalings.

**Left-hand side**
$$
n_t = \frac{k_+^4}{V_{+,0}^3}\,\hat n_{\hat t}.
$$

**Diffusion term**
$$
D_0 n_{xx}
= D_0\left(\frac{k_+}{V_{+,0}}\right)^2 \frac{k_+^3}{V_{+,0}^3}\,\hat n_{\hat x\hat x}
= \frac{k_+^4}{V_{+,0}^3}\left(\frac{D_0 k_+}{V_{+,0}^2}\right)\hat n_{\hat x\hat x}.
$$

**Reaction terms**
$$
-(k_+ + k_-)n
= -\frac{k_+^4}{V_{+,0}^3}\left(1+\frac{k_-}{k_+}\right)\hat n,
$$
$$
k_p n_+ = \frac{k_+^4}{V_{+,0}^3}\left(\frac{k_p}{k_+}\right)\hat n_+,\qquad
k_n n_- = \frac{k_+^4}{V_{+,0}^3}\left(\frac{k_n}{k_+}\right)\hat n_-.
$$

Now divide the whole equation by the common factor $\frac{k_+^4}{V_{+,0}^3}$. This gives:
$$
\hat n_{\hat t}
= \hat D_0\,\hat n_{\hat x\hat x}
-(1+\hat k_-)\hat n + \hat k_p\hat n_+ + \hat k_n\hat n_-.
$$

This is exactly the first line of the dimensionless system (5).

---

### 5) The other two equations (same idea)
Applying the same substitutions to the $+$ and $-$ equations gives:
$$
(\hat n_+)_{\hat t}+\left(\frac{V^+}{V_{+,0}}\hat n_+\right)_{\hat x}
= \hat n-\hat k_p\hat n_+,
$$
$$
(\hat n_-)_{\hat t}+\left(\frac{V^-}{V_{+,0}}\hat n_-\right)_{\hat x}
= \hat k_- \hat n-\hat k_n\hat n_-.
$$

---

### 6) Final dimensionless model
Therefore the dimensional model (1) is rewritten in the dimensionless form (5):
$$
\begin{aligned}
\hat n_{\hat t} &= \hat D_0\hat n_{\hat x\hat x}-(1+\hat k_-)\hat n+\hat k_p\hat n_+ + \hat k_n\hat n_-,\\
(\hat n_+)_{\hat t}+\left(\frac{V^+}{V_{+,0}}\hat n_+\right)_{\hat x} &= \hat n-\hat k_p\hat n_+,\\
(\hat n_-)_{\hat t}+\left(\frac{V^-}{V_{+,0}}\hat n_-\right)_{\hat x} &= \hat k_- \hat n-\hat k_n\hat n_-,
\qquad \hat x\in[0,\hat L].
\end{aligned}
$$

**Interpretation:** the dimensionless groups $\hat D_0,\hat k_-,\hat k_p,\hat k_n$ compare diffusion and reaction rates to the reference scales $V_{+,0}$ and $k_+$, and the velocities appear only as ratios $V^\pm/V_{+,0}$.


### Explanation of the initial and boundary conditions (simple)

At $t=0$ the free particles have a linear profile,
$$
n(x,0)=N_0+(N_L-N_0)\frac{x}{L},
$$
so the ends match $n(0,0)=N_0$ and $n(L,0)=N_L$. At the start there are no motor–bound particles inside the axon:
$$
n_+(x,0)=0,\qquad n_-(x,0)=0.
$$

For $t>0$ the free concentration is kept fixed at the boundaries,
$$
n(0,t)=N_0,\qquad n(L,t)=N_L,
$$
meaning both ends act like reservoirs. In addition, the left end loads particles onto $+$ motors and the right end loads particles onto $-$ motors:
$$
n_+(0,t)=N_0\sigma_0,\qquad n_-(L,t)=N_L\sigma_L.
$$
So, $+$-moving cargo enters from $x=0$ and $-$-moving cargo enters from $x=L$.


###  Discretization of the full model Eq. (13)

We discretize the spatial domain $[0,L]$ using a uniform grid
$$
x_i = i\,\Delta x, \qquad i = 0,1,\dots,N,
$$
where $\Delta x = L/N$.

Time is discretized as
$$
t^m = m\,\Delta t, \qquad m = 0,1,2,\dots .
$$

The discrete unknowns are defined by
$$
n_i^m \approx n(x_i,t^m), \qquad
n_{+,i}^m \approx n_+(x_i,t^m), \qquad
n_{-,i}^m \approx n_-(x_i,t^m).
$$

These quantities represent the numerical approximations of the solution at grid point $x_i$ and time level $t^m$.


### Step 2: Time discretization

The time derivatives in the system are approximated using a forward Euler scheme. 
For a generic variable $u(x,t)$, this gives
$$
\frac{\partial u}{\partial t}(x_i,t^m)
\approx
\frac{u_i^{m+1}-u_i^m}{\Delta t}.
$$

Applying this to the unknowns in the model yields
$$
\frac{n_i^{m+1}-n_i^m}{\Delta t}, \qquad
\frac{n_{+,i}^{m+1}-n_{+,i}^m}{\Delta t}, \qquad
\frac{n_{-,i}^{m+1}-n_{-,i}^m}{\Delta t}.
$$

This explicit time discretization updates the solution from time level $t^m$ to $t^{m+1}$.


### Step 3: Diffusion and reaction terms

#### Diffusion term (only in the equation for $n$)
The second derivative in space is approximated by a central difference:
$$
\frac{\partial^2 n}{\partial x^2}(x_i,t^m)
\approx
\frac{n_{i+1}^m - 2n_i^m + n_{i-1}^m}{(\Delta x)^2}.
$$
Hence,
$$
D_0\,n_{xx}(x_i,t^m)
\approx
D_0\,\frac{n_{i+1}^m - 2n_i^m + n_{i-1}^m}{(\Delta x)^2}.
$$

#### Reaction (source) terms
All reaction terms are local and are evaluated at the grid point $x_i$ at time level $t^m$.

For the $n$-equation:
$$
-(1+k_-)\,n + k_p n_+ + k_n n_-
\;\approx\;
-(1+k_-)\,n_i^m + k_p\,n_{+,i}^m + k_n\,n_{-,i}^m.
$$

For the $n_+$-equation (reaction part):
$$
n - k_p n_+
\;\approx\;
n_i^m - k_p\,n_{+,i}^m.
$$

For the $n_-$-equation (reaction part):
$$
k_- n - k_n n_-
\;\approx\;
k_-\,n_i^m - k_n\,n_{-,i}^m.
$$


### Step 4: Advection terms (upwind scheme)

The transport equations in (13) contain first–order advection terms.  
Following Task 2, we use an upwind discretization, where the spatial difference is chosen based on the direction of motion.

#### Advection for $n_+$ (transport to the right)
The $n_+$ equation has the form
$$
(n_+)_t + (n_+)_x = \text{(reaction/source)}.
$$
Since the characteristic speed is positive (flow to the right), we use a backward (left) difference:
$$
(n_+)_x(x_i,t^m) \approx \frac{n_{+,i}^m - n_{+,i-1}^m}{\Delta x}.
$$

#### Advection for $n_-$ (transport to the left)
The $n_-$ equation has the form
$$
(n_-)_t - (n_-)_x = \text{(reaction/source)}.
$$
This corresponds to advection with negative speed (flow to the left). Therefore we use a forward (right) difference:
$$
(n_-)_x(x_i,t^m) \approx \frac{n_{-,i+1}^m - n_{-,i}^m}{\Delta x}.
$$

With these choices, the advection discretization is consistent with the upwind principle: information is taken from the upstream direction.


### Step 5: Fully discrete scheme for the TJM (13)

Combining Step 2 (Forward Euler in time), Step 3 (diffusion + reactions), and Step 4 (upwind advection),
we obtain the following explicit update scheme for interior grid points $i=1,\dots,N-1$.

#### Equation for $n$
$$
\frac{n_i^{m+1}-n_i^m}{\Delta t}
=
D_0\,\frac{n_{i+1}^m-2n_i^m+n_{i-1}^m}{(\Delta x)^2}
-(1+k_-)\,n_i^m
+k_p\,n_{+,i}^m
+k_n\,n_{-,i}^m.
$$
Equivalently,
$$
n_i^{m+1}
=
n_i^m
+
\Delta t\left[
D_0\,\frac{n_{i+1}^m-2n_i^m+n_{i-1}^m}{(\Delta x)^2}
-(1+k_-)\,n_i^m
+k_p\,n_{+,i}^m
+k_n\,n_{-,i}^m
\right].
$$

#### Equation for $n_+$ (advection to the right: upwind/backward difference)
$$
\frac{n_{+,i}^{m+1}-n_{+,i}^m}{\Delta t}
+
\frac{n_{+,i}^m-n_{+,i-1}^m}{\Delta x}
=
n_i^m-k_p\,n_{+,i}^m.
$$
Equivalently,
$$
n_{+,i}^{m+1}
=
n_{+,i}^m
-
\frac{\Delta t}{\Delta x}\left(n_{+,i}^m-n_{+,i-1}^m\right)
+
\Delta t\left(n_i^m-k_p\,n_{+,i}^m\right).
$$

#### Equation for $n_-$ (advection to the left: upwind/forward difference)
$$
\frac{n_{-,i}^{m+1}-n_{-,i}^m}{\Delta t}
-
\frac{n_{-,i+1}^m-n_{-,i}^m}{\Delta x}
=
k_-\,n_i^m-k_n\,n_{-,i}^m.
$$
Equivalently,
$$
n_{-,i}^{m+1}
=
n_{-,i}^m
+
\frac{\Delta t}{\Delta x}\left(n_{-,i+1}^m-n_{-,i}^m\right)
+
\Delta t\left(k_-\,n_i^m-k_n\,n_{-,i}^m\right).
$$


### Step 6: Discrete initial and boundary conditions

The initial conditions are imposed at time level $m=0$.
For all spatial grid points $i=0,\dots,N$, we set
$$
n_i^0 = n(x_i,0), \qquad
n_{+,i}^0 = n_+(x_i,0), \qquad
n_{-,i}^0 = n_-(x_i,0),
$$
according to the prescribed initial profiles.

The boundary conditions are imposed at each time level $t^m$.
For the free particles, Dirichlet boundary conditions are applied:
$$
n_0^m = N_0, \qquad n_N^m = N_L.
$$

For the motor–bound particles, inflow boundary conditions are prescribed:
$$
n_{+,0}^m = N_0\,\sigma_0, \qquad
n_{-,N}^m = N_L\,\sigma_L.
$$

The discrete update scheme in Step 5 is applied at interior grid points
$i=1,\dots,N-1$, while the boundary values are fixed by the conditions above.


## Task 3(c): Model without source terms

We first consider the simplified traffic jam model obtained by setting the right-hand side of (13) to zero,
\[
\begin{aligned}
n_t &= D_0 n_{xx},\\
(n_+)_t + (n_+)_x &= 0,\\
(n_-)_t - (n_-)_x &= 0,
\end{aligned}
\]
which describes pure diffusion for the free particles and pure advection for motor–bound particles.

The equations are decoupled and are solved using an explicit diffusion scheme for \(n\) and upwind schemes
for \(n_+\) and \(n_-\). Parameters are taken from Table 1.



```python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# Task 3(c): model (14), RHS = 0
# n_t = D0 n_xx
# (n+)_t + (n+)_x = 0
# (n-)_t - (n-)_x = 0
# =========================

# ---- Table 1 parameters (dimensionless) ----
L   = 20.0
D0  = 0.4
N0  = 0.1
NL  = 0.01
sigma0 = 0.1
sigmaL = 0.1
dx  = 0.2

# Inflow values for motor-bound particles
nplus_in  = N0 * sigma0      # n_+(0,t)
nminus_in = NL * sigmaL      # n_-(L,t)

# ---- Grid ----
Nx = int(round(L/dx))                 # number of intervals
x  = np.linspace(0.0, L, Nx + 1)      # Nx+1 points: i=0..Nx

# ---- Stability constraints ----
# Advection speed magnitude is 1 in model (14)
dt_adv = dx                           # CFL: dt/dx <= 1  -> dt <= dx

# Explicit diffusion stability in 1D: dt <= dx^2/(2 D0)
dt_diff = dx**2 / (2.0 * D0)

# Choose a safe dt (smaller than both)
dt = 0.02  # safe choice; must satisfy dt <= min(dt_adv, dt_diff)

print(f"Nx = {Nx}, dx = {dx}")
print(f"dt_adv  = {dt_adv:.5f}, dt_diff = {dt_diff:.5f}, chosen dt = {dt:.5f}")
if dt > min(dt_adv, dt_diff):
    raise ValueError("dt is too large for stability. Choose a smaller dt.")

# ---- Simulation times ----
T_end = 20.0
T_compare = 10.0

Nt = int(np.ceil(T_end / dt))
t_grid = np.arange(Nt + 1) * dt

# Helper: nearest time index to a requested time
def tidx(t):
    return int(round(t / dt))

idx0  = tidx(0.0)
idx10 = tidx(10.0)
idx20 = tidx(20.0)

# ---- Initial conditions ----
# n(x,0) linear profile
n = N0 + (NL - N0) * (x / L)

# n+(x,0) = 0, n-(x,0) = 0
nplus  = np.zeros_like(x)
nminus = np.zeros_like(x)

# ---- Boundary conditions for all t ----
def apply_bc(n, nplus, nminus):
    # Dirichlet for n
    n[0]  = N0
    n[-1] = NL
    # Inflow BCs for motor-bound
    nplus[0]  = nplus_in      # inflow at left
    nminus[-1] = nminus_in    # inflow at right
    return n, nplus, nminus

n, nplus, nminus = apply_bc(n, nplus, nminus)

# ---- Storage for snapshots ----
snapshots = {}

# ---- Time stepping ----
r = D0 * dt / dx**2      # diffusion number
lam = dt / dx            # CFL number for advection (speed 1)

for m in range(Nt + 1):

    # Save snapshots at requested times
    if m in [idx0, idx10, idx20]:
        snapshots[m*dt] = (n.copy(), nplus.copy(), nminus.copy())

    # stop after final time stored
    if m == Nt:
        break

    n_old     = n.copy()
    nplus_old = nplus.copy()
    nminus_old= nminus.copy()

    # ---- Update n (diffusion) for interior points ----
    # n_i^{m+1} = n_i^m + D0*dt*(n_{i+1}-2n_i+n_{i-1})/dx^2
    n[1:-1] = n_old[1:-1] + r * (n_old[2:] - 2.0*n_old[1:-1] + n_old[:-2])

    # ---- Update n+ (advection right) using upwind (backward difference) ----
    # (n+)_t + (n+)_x = 0
    # n+_i^{m+1} = n+_i^m - (dt/dx)*(n+_i^m - n+_{i-1}^m)
    nplus[1:] = nplus_old[1:] - lam * (nplus_old[1:] - nplus_old[:-1])

    # ---- Update n- (advection left) using upwind for negative speed (forward difference) ----
    # (n-)_t - (n-)_x = 0  <=>  (n-)_t + (-1)(n-)_x = 0
    # n-_i^{m+1} = n-_i^m + (dt/dx)*(n-_{i+1}^m - n-_i^m)
    nminus[:-1] = nminus_old[:-1] + lam * (nminus_old[1:] - nminus_old[:-1])

    # ---- Apply boundary conditions ----
    n, nplus, nminus = apply_bc(n, nplus, nminus)

# ---- Exact solutions at T=10 for comparison ----
T = T_compare

# Exact n: stays linear for all t
n_exact = N0 + (NL - N0) * (x / L)

# Exact n+: step moving right with speed 1, inflow constant
nplus_exact = np.where(x < T, nplus_in, 0.0)

# Exact n-: step moving left with speed 1, inflow at right constant
nminus_exact = np.where(x > (L - T), nminus_in, 0.0)

# ---- Plot snapshots ----
def plot_snapshot(t, n_num, np_num, nm_num, overlay_exact=False):
    plt.figure()
    plt.plot(x, n_num, label="n (numerical)")
    plt.plot(x, np_num, label="n+ (numerical)")
    plt.plot(x, nm_num, label="n- (numerical)")

    if overlay_exact:
        plt.plot(x, n_exact, label="n (exact)")
        plt.plot(x, nplus_exact, label="n+ (exact)")
        plt.plot(x, nminus_exact, label="n- (exact)")

    plt.title(f"Model (14), RHS=0 at t = {t:.2f}")
    plt.xlabel("x")
    plt.ylabel("concentration")
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot at t=0,10,20
for t in [0.0, 10.0, 20.0]:
    # Find the stored snapshot time closest to requested t
    t_store = min(snapshots.keys(), key=lambda s: abs(s - t))
    n_s, np_s, nm_s = snapshots[t_store]
    plot_snapshot(t_store, n_s, np_s, nm_s, overlay_exact=(abs(t_store-10.0) < 1e-12))

print("Done.")
```

### Discussion of results

The free particle concentration \(n\) remains close to the initial linear profile for all times, in agreement
with the exact solution of the diffusion equation.

The concentrations \(n_+\) and \(n_-\) form traveling fronts entering from \(x=0\) and \(x=L\), respectively.
The fronts propagate with unit speed, so that the characteristic travel time across the domain is \(T \approx L\).

At \(T=10\) the numerical solutions agree well with the exact step–front solutions. The numerical fronts are
slightly smeared due to numerical diffusion introduced by the upwind scheme, which is expected.


### Grid refinement test (convergence at T=10)


```python
import numpy as np
import pandas as pd

# =========================
# Convergence test for Task 3(c), model (14), RHS=0
# Compare numerical vs exact at T=10 for different dx.
# =========================

# ---- Table 1 parameters ----
L   = 20.0
D0  = 0.4
N0  = 0.1
NL  = 0.01
sigma0 = 0.1
sigmaL = 0.1

# Inflow values
nplus_in  = N0 * sigma0   # 0.01
nminus_in = NL * sigmaL   # 0.001

T = 10.0  # comparison time

def exact_solutions(x, T):
    # Exact n: stays linear
    n_ex = N0 + (NL - N0) * (x / L)
    # Exact n+: step front at x = T
    np_ex = np.where(x < T, nplus_in, 0.0)
    # Exact n-: step front at x = L - T
    nm_ex = np.where(x > (L - T), nminus_in, 0.0)
    return n_ex, np_ex, nm_ex

def stable_dt(dx, D0, cfl_safety=0.9, diff_safety=0.9):
    # Advection stability (speed=1): dt <= dx
    dt_adv  = cfl_safety * dx
    # Diffusion explicit stability: dt <= dx^2/(2D0)
    dt_diff = diff_safety * (dx**2) / (2.0 * D0)
    return min(dt_adv, dt_diff)

def run_model14_to_T(dx, dt, T):
    # Grid
    Nx = int(round(L/dx))
    x  = np.linspace(0.0, L, Nx+1)

    # Initial conditions
    n = N0 + (NL - N0) * (x / L)
    nplus  = np.zeros_like(x)
    nminus = np.zeros_like(x)

    # Apply BCs
    def apply_bc(n, nplus, nminus):
        n[0]  = N0
        n[-1] = NL
        nplus[0]   = nplus_in
        nminus[-1] = nminus_in
        return n, nplus, nminus

    n, nplus, nminus = apply_bc(n, nplus, nminus)

    # Precompute constants
    r   = D0 * dt / dx**2
    lam = dt / dx

    # Time steps
    Nt = int(np.ceil(T / dt))
    for m in range(Nt):
        n_old     = n.copy()
        np_old    = nplus.copy()
        nm_old    = nminus.copy()

        # Diffusion update (interior)
        n[1:-1] = n_old[1:-1] + r * (n_old[2:] - 2.0*n_old[1:-1] + n_old[:-2])

        # Advection right for n+ (upwind)
        nplus[1:] = np_old[1:] - lam * (np_old[1:] - np_old[:-1])

        # Advection left for n- (upwind for negative speed)
        nminus[:-1] = nm_old[:-1] + lam * (nm_old[1:] - nm_old[:-1])

        # BCs
        n, nplus, nminus = apply_bc(n, nplus, nminus)

    # Effective time reached
    t_reached = Nt * dt
    return x, n, nplus, nminus, t_reached

def error_metrics(num, ex):
    # max-norm and L2 discrete norm
    e = num - ex
    e_inf = np.max(np.abs(e))
    e_l2  = np.sqrt(np.mean(e**2))
    return e_inf, e_l2


dx_list = [0.2, 0.1, 0.05]
rows = []

for dx in dx_list:
    dt = stable_dt(dx, D0, cfl_safety=0.9, diff_safety=0.9)

    x, n_num, np_num, nm_num, t_reached = run_model14_to_T(dx, dt, T)

    # Exact at T (we compare at the intended T; note t_reached is very close)
    n_ex, np_ex, nm_ex = exact_solutions(x, T)

    # Errors
    n_inf,  n_l2  = error_metrics(n_num,  n_ex)
    np_inf, np_l2 = error_metrics(np_num, np_ex)
    nm_inf, nm_l2 = error_metrics(nm_num, nm_ex)

    rows.append({
        "dx": dx,
        "dt": dt,
        "Nt": int(np.ceil(T/dt)),
        "t_reached": t_reached,
        "err_inf(n)":  n_inf,
        "err_L2(n)":   n_l2,
        "err_inf(n+)": np_inf,
        "err_L2(n+)":  np_l2,
        "err_inf(n-)": nm_inf,
        "err_L2(n-)":  nm_l2,
    })

df = pd.DataFrame(rows)
df
```

### Answers to the questions (RHS = 0, model (14))

**1) Behavior generated by the model.**  
The simplified system (14) is decoupled. The free particles satisfy a pure diffusion equation, while
$n_+$ and $n_-$ satisfy pure advection equations with speeds $+1$ and $-1$. Numerically, this produces:
- a diffusion profile for $n$ (here it remains close to the initial linear profile),
- a right–moving front for $n_+$ entering from $x=0$,
- a left–moving front for $n_-$ entering from $x=L$.
The upwind scheme introduces slight numerical diffusion, so the fronts are smeared.

**2) Suitable simulation period for fronts to propagate end-to-end.**  
Since the advection speeds are $\pm 1$, the travel time across the domain is
$$
T_{\text{travel}} \approx \frac{L}{1} = L.
$$
With $L=20$, a suitable simulation period is therefore $T \gtrsim 20$.

**3) Exact solutions and comparison at $T=10$.**  
For the diffusion equation with linear boundary data and linear initial profile,
$$
n(x,t)=N_0+(N_L-N_0)\frac{x}{L}.
$$
For the advection equations with constant inflow and zero initial data,
$$
n_+(x,t)=
\begin{cases}
N_0\sigma_0, & x<t,\\
0, & x>t,
\end{cases}
\qquad
n_-(x,t)=
\begin{cases}
N_L\sigma_L, & x>L-t,\\
0, & x<L-t.
\end{cases}
$$
At $T=10$ the fronts are located near $x=10$. The numerical solutions agree well with these exact
solutions, except for slight front smearing due to upwind discretization.

**4) Convergence and stability.**  
Convergence was checked by refining the grid and comparing with the exact solutions at $T=10$.
The schemes must satisfy:
- advection (CFL): $\Delta t/\Delta x \le 1$,
- diffusion (explicit): $\Delta t \le (\Delta x)^2/(2D_0)$.

**5) Summing the equations in the full model (13).**  
Let the total mass be $M = n + n_+ + n_-$. Summing the equations in (13) yields a conservation law
$$
M_t + J_x = 0,
$$
which reflects conservation of total particle mass (changes occur through boundary fluxes).

**6) Time to reach steady state of the total mass.**  
A typical time scale is the maximum of the advection time and diffusion time,
$$
T_{\text{steady}} \sim \max\!\left(L,\ \frac{L^2}{D_0}\right).
$$
With $L=20$ and $D_0=0.4$, this gives $L^2/D_0 = 1000$, so diffusion acts on a much longer time scale
than advection. In the present RHS=0 test, $n$ is already in its steady linear profile and the main
transient behavior is the front propagation of $n_+$ and $n_-$ on the scale $T\sim L$.

**7) Total flux and whether it is constant.**  
The total flux is
$$
J = n_+ - n_- - D_0 n_x.
$$
From $M_t + J_x = 0$, we have that during transients $J$ generally varies with $x$, while at steady
state $M_t=0$ implies $J_x=0$, so $J$ becomes constant in space.


## Task 3(d): Reduced submodel (15)

We include attachment/detachment effects stepwise and study the simplified system

$$
n_t = D_0\,n_{xx} - (1+k_-)\,n, \qquad x\in[0,L],
$$

$$
(n_+)_t + (n_+)_x = n,
$$

$$
(n_-)_t - (n_-)_x = k_-\,n.
\tag{15}
$$

We use the same initial and boundary conditions as in the previous step, and simulate on the time interval
$$
t \in [0,T], \qquad T=30.
$$

We also monitor the (discrete) total flux
$$
J(x,t) = n_+(x,t) - n_-(x,t) - D_0\,n_x(x,t).
$$

### What effects are accounted for?
- $D_0\,n_{xx}$: diffusion of free particles $n$.
- $-(1+k_-)\,n$: removal of free particles due to attachment into motor–bound states.
- RHS in the transport equations: creation of motor–bound particles from the free pool:
  - $n$ creates $n_+$,
  - $k_-\,n$ creates $n_-$.

### What happens at the very beginning of the simulation?
Initially $n_+(x,0)=0$ and $n_-(x,0)=0$ in the interior, but since the source terms depend on $n(x,0)$,
motor–bound particles are created immediately wherever $n(x,0)>0$. In addition, the boundary inflow conditions
inject $n_+$ at $x=0$ and $n_-$ at $x=L$ for all $t>0$.


```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------
# Parameters (Table 1)
# -------------------------
L   = 20.0
D0  = 0.4
N0  = 0.1
NL  = 0.01
kminus = 1.0
sigma0 = 0.1
sigmaL = 0.1
dx  = 0.2

nplus_in  = N0 * sigma0      # n_+(0,t)
nminus_in = NL * sigmaL      # n_-(L,t)

# -------------------------
# Grid
# -------------------------
Nx = int(round(L/dx))
x  = np.linspace(0.0, L, Nx+1)

# -------------------------
# Time step (stability)
# Advection CFL: dt/dx <= 1
# Diffusion explicit: dt <= dx^2/(2D0)
# Reaction: explicit stable if dt*(1+kminus) not too large (we keep dt small anyway)
# -------------------------
dt_adv  = dx
dt_diff = dx**2/(2*D0)
dt = 0.02  # safe
print(f"dx={dx}, dt_adv={dt_adv:.4f}, dt_diff={dt_diff:.4f}, chosen dt={dt:.4f}")
if dt > min(dt_adv, dt_diff):
    raise ValueError("dt too large for stability. Choose smaller dt.")

# -------------------------
# Simulation time
# -------------------------
T_end = 30.0
Nt = int(np.ceil(T_end/dt))

# -------------------------
# Initial conditions
# -------------------------
n = N0 + (NL - N0) * (x/L)        # linear profile
nplus  = np.zeros_like(x)
nminus = np.zeros_like(x)

def apply_bc(n, nplus, nminus):
    n[0]  = N0
    n[-1] = NL
    nplus[0]   = nplus_in
    nminus[-1] = nminus_in
    return n, nplus, nminus

n, nplus, nminus = apply_bc(n, nplus, nminus)

r   = D0*dt/dx**2
lam = dt/dx

# -------------------------
# Discrete flux J = n+ - n- - D0 n_x
# -------------------------
def compute_J(n, nplus, nminus):
    nx = np.zeros_like(n)
    nx[1:-1] = (n[2:] - n[:-2])/(2*dx)
    nx[0]    = (n[1] - n[0])/dx
    nx[-1]   = (n[-1] - n[-2])/dx
    return nplus - nminus - D0*nx

# Store snapshots
snap_times = [0.0, 29.0, 30.0]
snap_idx = {int(round(t/dt)): t for t in snap_times}
snaps = {}

# -------------------------
# Time loop (model 15)
# n_t = D0 n_xx - (1+kminus)n
# (n+)_t + (n+)_x = n
# (n-)_t - (n-)_x = kminus*n
# -------------------------
for m in range(Nt+1):

    if m in snap_idx:
        t = snap_idx[m]
        snaps[t] = (n.copy(), nplus.copy(), nminus.copy(), compute_J(n, nplus, nminus))

    if m == Nt:
        break

    n_old  = n.copy()
    np_old = nplus.copy()
    nm_old = nminus.copy()

    # Update n: diffusion + decay
    n[1:-1] = (n_old[1:-1]
               + r*(n_old[2:] - 2*n_old[1:-1] + n_old[:-2])
               - dt*(1.0 + kminus)*n_old[1:-1])

    # Update n+: upwind (right) + source n
    nplus[1:] = (np_old[1:]
                 - lam*(np_old[1:] - np_old[:-1])
                 + dt*n_old[1:])

    # Update n-: upwind for negative speed (left) + source kminus*n
    nminus[:-1] = (nm_old[:-1]
                    + lam*(nm_old[1:] - nm_old[:-1])
                    + dt*kminus*n_old[:-1])

    # Apply BCs
    n, nplus, nminus = apply_bc(n, nplus, nminus)

# -------------------------
# Plots: n, n+, n- at snapshots
# -------------------------
for t in snap_times:
    n_s, np_s, nm_s, J_s = snaps[t]
    plt.figure()
    plt.plot(x, n_s, label="n")
    plt.plot(x, np_s, label="n+")
    plt.plot(x, nm_s, label="n-")
    plt.title(f"Model (15) at t = {t:g}")
    plt.xlabel("x")
    plt.ylabel("concentration")
    plt.grid(True)
    plt.legend()
    plt.show()

# -------------------------
# Plots: flux J(x,t) at snapshots
# -------------------------
for t in snap_times:
    n_s, np_s, nm_s, J_s = snaps[t]
    plt.figure()
    plt.plot(x, J_s, label="J")
    plt.title(f"Flux J(x,t) = n+ - n- - D0 n_x at t = {t:g}")
    plt.xlabel("x")
    plt.ylabel("J")
    plt.grid(True)
    plt.legend()
    plt.show()
```

### Behavior of the total flux

The discrete flux
$$
J(x,t)=n_+(x,t)-n_-(x,t)-D_0\,n_x(x,t)
$$
varies strongly in space during the initial transient period. This reflects the local production of
motor–bound particles due to attachment and the redistribution caused by advection and diffusion.

As time increases, the solution approaches a steady regime. In this limit $M_t \approx 0$ for
$M=n+n_++n_-$, and the conservation law $M_t+J_x=0$ implies $J_x\approx0$. Consequently,
the flux becomes nearly constant in space, which is clearly observed in the numerical results at $t=30$.

### Steady-state check for the flux $J(x,t)$ (Task 3(d))

To verify that the flux profile has reached a steady (time-independent) state, we compare the flux at two late times.

The flux is defined as
$$
J(x,t) = n_+(x,t) - n_-(x,t) - D_0\,\frac{\partial n}{\partial x}(x,t).
$$

We compute the maximum difference between the two flux profiles:
$$
\max_{x \in [0,L]} \left| J(x,30) - J(x,29) \right|
= 5.390431783436345 \times 10^{-7}.
$$

Since this value is extremely small, we conclude that $J(x,t)$ is essentially steady by $t = 30$.


```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------
# Parameters
# -------------------------
L   = 20.0
D0  = 0.4
N0  = 0.1
NL  = 0.01
sigma0 = 0.1
sigmaL = 0.1
dx  = 0.2

# detachment rates 
kp = 0.5
kn = 0.5

# inflow boundary values for bound states
nplus_in  = N0 * sigma0      # n_+(0,t)
nminus_in = NL * sigmaL      # n_-(L,t)

# -------------------------
# Grid
# -------------------------
Nx = int(round(L/dx))
x  = np.linspace(0.0, L, Nx+1)

# -------------------------
# Time step (stability)
# -------------------------
dt_adv  = dx
dt_diff = dx**2/(2*D0)
dt_reac = 1.0/max(kp, kn)
dt = 0.02

print(f"dx={dx}, dt_adv={dt_adv:.4f}, dt_diff={dt_diff:.4f}, dt_reac={dt_reac:.4f}, chosen dt={dt:.4f}")
if dt > min(dt_adv, dt_diff, dt_reac):
    raise ValueError("dt too large for stability. Choose smaller dt.")

# -------------------------
# Simulation time
# -------------------------
T_end = 200.0
Nt = int(np.ceil(T_end/dt))

# -------------------------
# Initial conditions
# -------------------------
n = N0 + (NL - N0) * (x/L)        # linear profile
nplus  = np.zeros_like(x)
nminus = np.zeros_like(x)

def apply_bc(n, nplus, nminus):
    n[0]  = N0
    n[-1] = NL
    nplus[0]   = nplus_in
    nminus[-1] = nminus_in
    return n, nplus, nminus

n, nplus, nminus = apply_bc(n, nplus, nminus)

r   = D0*dt/dx**2
lam = dt/dx

# -------------------------
# Flux J = n+ - n- - D0 n_x
# -------------------------
def compute_J(n, nplus, nminus):
    nx = np.zeros_like(n)
    nx[1:-1] = (n[2:] - n[:-2])/(2*dx)
    nx[0]    = (n[1] - n[0])/dx
    nx[-1]   = (n[-1] - n[-2])/dx
    return nplus - nminus - D0*nx

# -------------------------
# Snapshots
# -------------------------
snap_times = [0.0, 50.0, 100.0, 199.00, 200.0]
snap_idx = {int(t/dt + 1e-12): t for t in snap_times}  # safer than round
snaps = {}

# -------------------------
# Time loop (model 16)
# n_t = D0 n_xx + kp n+ + kn n-
# (n+)_t + (n+)_x = -kp n+
# (n-)_t - (n-)_x = -kn n-
# -------------------------
for m in range(Nt+1):

    if m in snap_idx:
        t = snap_idx[m]
        snaps[t] = (n.copy(), nplus.copy(), nminus.copy(), compute_J(n, nplus, nminus).copy())

    if m == Nt:
        break

    n_old  = n.copy()
    np_old = nplus.copy()
    nm_old = nminus.copy()

    # Update n: diffusion + gain from detachment
    n[1:-1] = (n_old[1:-1]
               + r*(n_old[2:] - 2*n_old[1:-1] + n_old[:-2])
               + dt*(kp*np_old[1:-1] + kn*nm_old[1:-1]))

    # Update n+: upwind right + decay
    nplus[1:] = (np_old[1:]
                 - lam*(np_old[1:] - np_old[:-1])
                 - dt*kp*np_old[1:])

    # Update n-: upwind left + decay
    nminus[:-1] = (nm_old[:-1]
                   + lam*(nm_old[1:] - nm_old[:-1])
                   - dt*kn*nm_old[:-1])

    # Apply BCs
    n, nplus, nminus = apply_bc(n, nplus, nminus)

print("Saved snapshots:", sorted(snaps.keys()))

# -------------------------
# Plot ONLY t=0 and t=200 (report style)
# -------------------------
tA, tB = 0.0, 200.0
if tA not in snaps or tB not in snaps:
    raise KeyError("Snapshots missing. Check 'Saved snapshots' print above.")

n0, np0, nm0, J0 = snaps[tA]
n1, np1, nm1, J1 = snaps[tB]

plt.figure(figsize=(6,4))
plt.plot(x, n0,  "--", label="n (t=0)")
plt.plot(x, np0, "--", label="n+ (t=0)")
plt.plot(x, nm0, "--", label="n- (t=0)")
plt.plot(x, n1,  "-", label="n (t=200)")
plt.plot(x, np1, "-", label="n+ (t=200)")
plt.plot(x, nm1, "-", label="n- (t=200)")
plt.title("Model (16): densities")
plt.xlabel("x"); plt.ylabel("concentration")
plt.grid(True); plt.legend(fontsize=8)
plt.show()

plt.figure(figsize=(6,4))
plt.plot(x, J0, "--", label="J (t=0)")
plt.plot(x, J1, "-",  label="J (t=200)")
plt.title("Model (16): flux J(x,t)")
plt.xlabel("x"); plt.ylabel("J")
plt.grid(True); plt.legend()
plt.show()
```

## Task 3(e) — Detachment only (model (16))

### What effects are accounted for?
Model (16) includes:
- diffusion of free particles $n$ (term $D_0\,n_{xx}$),
- directed transport of motor-bound particles $n_+$ (to the right) and $n_-$ (to the left),
- detachment from motors: $n_+$ and $n_-$ decay with rates $k_p$ and $k_n$, and detached particles become free (source $k_p n_+ + k_n n_-$ in the $n$-equation).
There is **no attachment** $n \to n_\pm$.

The equations are:
$$
n_t = D_0\,n_{xx} + k_p n_+ + k_n n_-,
$$
$$
(n_+)_t + (n_+)_x = -k_p n_+,
\qquad
(n_-)_t - (n_-)_x = -k_n n_-.
$$

### How long to run to get a steady flux $J(x,t)$?
We compute the flux as:
$$
J(x,t)=n_+(x,t)-n_-(x,t)-D_0\,\frac{\partial n}{\partial x}(x,t).
$$

A steady profile is reached by $t=200$, since:
$$
\max_{x\in[0,L]} \left|J(x,200)-J(x,199)\right|
= 3.13508455029865\times 10^{-6}.
$$

### Description of the solution (in words)
Because attachment is not included, $n_+$ and $n_-$ enter only from the boundaries and then decay while moving along the axon. Therefore $n_+$ is mainly near $x=0$ and $n_-$ mainly near $x=L$. Detachment transfers mass to the free state $n$, which spreads by diffusion, giving a smooth steady flux $J(x)$.


### Task 3(f): Healthy vs unhealthy transport (full system)

We run the full system (13), which includes both attachment (free particles bind to motors) and detachment (motor-bound particles fall off and become free). A “healthy” transport corresponds to bounded, smooth profiles of $n$, $n_+$ and $n_-$ and a flux $J(x,t)$ that is approximately stable and does not collapse locally. In contrast, an “unhealthy” state (jamming) would be characterized by a strong local accumulation (sharp peak) in the densities and a significant local reduction of the flux, indicating that cargo gets stuck and forms a traffic jam.

```python
import numpy as np
import matplotlib.pyplot as plt

# ============================================================
# Hypothesis I (g): velocity depends on n+ and n-
# f(u)= +/- u exp(-A u)
# Flux discretization (19)-(20) with M=1
# ============================================================

# -------------------------
# Parameters 
# -------------------------
L   = 20.0
D0  = 0.4
N0  = 0.1
NL  = 0.01

sigma0 = 0.1
sigmaL = 0.1

kminus = 1.0   # attachment ratio into n-
kp = 0.5       # detachment from n+
kn = 0.5       # detachment from n-

dx = 0.2
V0 = 1.0       # V_{+,0} from (12);

# boundary inflow for bound states
nplus_in  = N0 * sigma0      # n_+(0,t)
nminus_in = NL * sigmaL      # n_-(L,t)

# grid
Nx = int(round(L/dx))
x  = np.linspace(0.0, L, Nx+1)

# stable dt: diffusion + LF flux (M=1) + reactions
dt_diff = dx**2/(2*D0)
dt_adv  = dx/1.0          # since M=1 in numerical flux, use dt <= dx
dt_reac = 1.0/max(1.0 + kminus, kp, kn)
dt = 0.02

print(f"dx={dx}, dt_diff={dt_diff:.4f}, dt_adv={dt_adv:.4f}, dt_reac={dt_reac:.4f}, chosen dt={dt:.4f}")
if dt > min(dt_diff, dt_adv, dt_reac):
    raise ValueError("dt too large for stability; choose smaller dt.")

r = D0*dt/dx**2

def apply_bc(n, nplus, nminus):
    n[0]  = N0
    n[-1] = NL
    nplus[0]   = nplus_in
    nminus[-1] = nminus_in
    return n, nplus, nminus

def compute_nx(n):
    nx = np.zeros_like(n)
    nx[1:-1] = (n[2:] - n[:-2])/(2*dx)
    nx[0]    = (n[1] - n[0])/dx
    nx[-1]   = (n[-1] - n[-2])/dx
    return nx

# Flux J requested in the task text:
# J(x,t) = n+ exp(-A n+) - n- exp(-A n-) + D0 n_x   (as in your screenshot)
# (If you want your old sign, change +D0*nx to -D0*nx.)
def compute_J(n, nplus, nminus, A):
    nx = compute_nx(n)
    return V0*(nplus*np.exp(-A*nplus) - nminus*np.exp(-A*nminus)) + D0*nx

# --------- Numerical flux (19)-(20), with M=1 ----------
Mnum = 1.0

def advect_update(u, A, sign):
    """
    Conservative update for u_t + (f(u))_x = 0 using (19)-(20)
    sign = +1 for n+  => f(u)= +V0*u*exp(-A u)
    sign = -1 for n-  => f(u)= -V0*u*exp(-A u)
    """
    f = sign * V0 * u * np.exp(-A*u)

    # F_{j+1/2} for j=0..Nx-1  (interfaces between u[j] and u[j+1])
    F_half = 0.5*(f[:-1] + f[1:]) - 0.5*Mnum*(u[1:] - u[:-1])

    u_new = u.copy()
    # u_j^{n+1} = u_j^n - dt/dx (F_{j+1/2} - F_{j-1/2})
    u_new[1:-1] = u[1:-1] - (dt/dx)*(F_half[1:] - F_half[:-1])

    return u_new

# ============================================================
# 1) Plot V+(u) and V-(u) for some A values
# ============================================================
u = np.linspace(0, 0.12, 300)
A_list_vel = [0, 2, 5, 7, 10]

plt.figure(figsize=(6,4))
for A in A_list_vel:
    plt.plot(u, V0*np.exp(-A*u), label=f"A={A}")
plt.title("Velocity factor exp(-A u) (for V+(n+))")
plt.xlabel("u"); plt.ylabel("exp(-A u)")
plt.grid(True); plt.legend()
plt.show()

plt.figure(figsize=(6,4))
for A in A_list_vel:
    plt.plot(u, -V0*np.exp(-A*u), label=f"A={A}")
plt.title("Velocity V-(n-) = -V0 exp(-A u)")
plt.xlabel("u"); plt.ylabel("V-")
plt.grid(True); plt.legend()
plt.show()

# ============================================================
# 2) Run simulations for A = 0, 2, 7 and compare
# ============================================================
A_cases = [0, 2, 7]
T_end = 100.0
Nt = int(np.ceil(T_end/dt))

# time series locations
x_center = 10.0
j_center = int(round(x_center/dx))

results = {}

for A in A_cases:

    # initial conditions
    n = N0 + (NL - N0)*(x/L)
    nplus = np.zeros_like(x)
    nminus = np.zeros_like(x)
    n, nplus, nminus = apply_bc(n, nplus, nminus)

    J_center = []
    J_avg = []
    tgrid = []

    for m in range(Nt+1):

        # record J
        J = compute_J(n, nplus, nminus, A)
        J_center.append(J[j_center])
        J_avg.append(np.trapezoid(J, x)/L)
        tgrid.append(m*dt)

        if m == Nt:
            break

        n_old = n.copy()
        np_old = nplus.copy()
        nm_old = nminus.copy()

        # ---- n: diffusion + attachment sink + detachment source (same as full system)
        n[1:-1] = (n_old[1:-1]
                   + r*(n_old[2:] - 2*n_old[1:-1] + n_old[:-2])
                   - dt*(1.0 + kminus)*n_old[1:-1]
                   + dt*(kp*np_old[1:-1] + kn*nm_old[1:-1]))

        # ---- n+ advection with new flux + source - decay
        nplus_adv = advect_update(np_old, A=A, sign=+1)
        nplus = nplus_adv
        nplus[1:-1] = nplus[1:-1] + dt*(n_old[1:-1] - kp*np_old[1:-1])

        # ---- n- advection with new flux + source - decay
        nminus_adv = advect_update(nm_old, A=A, sign=-1)
        nminus = nminus_adv
        nminus[1:-1] = nminus[1:-1] + dt*(kminus*n_old[1:-1] - kn*nm_old[1:-1])

        # BCs
        n, nplus, nminus = apply_bc(n, nplus, nminus)

        # safety: avoid negative values from numerics
        n = np.maximum(n, 0.0)
        nplus = np.maximum(nplus, 0.0)
        nminus = np.maximum(nminus, 0.0)

    # final profiles + peaks
    J_final = compute_J(n, nplus, nminus, A)
    results[A] = {
        "n": n, "nplus": nplus, "nminus": nminus, "J": J_final,
        "t": np.array(tgrid), "J_center": np.array(J_center), "J_avg": np.array(J_avg),
        "peak_nplus": float(np.max(nplus)), "peak_nminus": float(np.max(nminus))
    }

# ============================================================
# 3) Plots "similar to Fig 2": final profiles for different A
# ============================================================
plt.figure(figsize=(7,4))
for A in A_cases:
    plt.plot(x, results[A]["nplus"], label=f"n+ (A={A})")
plt.title("Final n+ profile for different A")
plt.xlabel("x"); plt.ylabel("n+")
plt.grid(True); plt.legend()
plt.show()

plt.figure(figsize=(7,4))
for A in A_cases:
    plt.plot(x, results[A]["nminus"], label=f"n- (A={A})")
plt.title("Final n- profile for different A")
plt.xlabel("x"); plt.ylabel("n-")
plt.grid(True); plt.legend()
plt.show()

plt.figure(figsize=(7,4))
for A in A_cases:
    plt.plot(x, results[A]["n"], label=f"n (A={A})")
plt.title("Final n profile for different A")
plt.xlabel("x"); plt.ylabel("n")
plt.grid(True); plt.legend()
plt.show()

plt.figure(figsize=(7,4))
for A in A_cases:
    plt.plot(x, results[A]["J"], label=f"J (A={A})")
plt.title("Final flux J(x,T) for different A")
plt.xlabel("x"); plt.ylabel("J")
plt.grid(True); plt.legend()
plt.show()

# ============================================================
# 4) Transport reduction: J(x=10,t) and averaged J(t)
# ============================================================
plt.figure(figsize=(7,4))
for A in A_cases:
    plt.plot(results[A]["t"], results[A]["J_center"], label=f"A={A}")
plt.title("J(x=10,t) for different A")
plt.xlabel("t"); plt.ylabel("J(x=10,t)")
plt.grid(True); plt.legend()
plt.show()

plt.figure(figsize=(7,4))
for A in A_cases:
    plt.plot(results[A]["t"], results[A]["J_avg"], label=f"A={A}")
plt.title("Averaged flux  J̄(t) = (1/L)∫ J(x,t) dx")
plt.xlabel("t"); plt.ylabel("J̄(t)")
plt.grid(True); plt.legend()
plt.show()

# ============================================================
# 5) Peak values (jamming indicators)
# ============================================================
for A in A_cases:
    print(f"A={A:>2}: peak(n+)={results[A]['peak_nplus']:.6f}, peak(n-)={results[A]['peak_nminus']:.6f}")
```

## Task 3(g) — Hypothesis I (density-dependent motor velocity)

Hypothesis I assumes that motor velocities decrease when the density of motor-bound particles increases:
$$
V_+(n_+) = V_{+,0}\,e^{-A n_+}, \qquad
V_-(n_-) = -V_{+,0}\,e^{-A n_-}.
$$
Thus, larger $A$ produces stronger slowing of transport for the same densities.

### Effect on densities (jamming tendency)
We simulated $A=0,2,7$. The peak values (indicating potential accumulation/jamming) increase with $A$:
- $A=0$: $\max_x n_+ = 0.040921$, $\max_x n_- = 0.059679$
- $A=2$: $\max_x n_+ = 0.044001$, $\max_x n_- = 0.062182$
- $A=7$: $\max_x n_+ = 0.056580$, $\max_x n_- = 0.073114$

The profiles show a more pronounced local build-up near the left boundary for $n_+$ and near the right boundary for $n_-$ when $A$ is large, consistent with a jamming mechanism.

### Effect on transport
The transport is computed as
$$
J(x,t)= n_+(x,t)\,e^{-A n_+(x,t)} - n_-(x,t)\,e^{-A n_-(x,t)} + D_0\,n_x(x,t).
$$
Both the flux at the center $J(x=10,t)$ and the spatially averaged flux
$$
\bar J(t) = \frac{1}{L}\int_0^L J(x,t)\,dx
$$
decrease as $A$ increases. Therefore, higher $A$ (stronger density-dependent slowing) reduces total transport and increases accumulation, which indicates a stronger jamming effect compared to $A=0$.

- We studied a transport model for axonal cargo with three populations: free particles $n$ (diffusion) and motor–bound particles $n_+$ and $n_-$ (directed transport in $\pm$ directions), coupled by attachment and detachment.

- Using an explicit scheme (diffusion + advection + reaction terms), the full system reaches a steady transport regime. In our healthy baseline case, the density profiles remain bounded and the flux becomes essentially time–independent.

- **Detachment-only model:** motor–bound particles decay while moving, and mass is transferred back to the free state. The flux became steady by $t=200$ (very small change between late times).

- **Full system (attachment + detachment):** steady, bounded profiles are obtained for $n$, $n_+$, and $n_-$, and the flux $J(x,t)$ approaches an almost constant profile in the interior. This corresponds to a “healthy” transport without strong local accumulation.

- **Hypothesis I (density-dependent velocity):** increasing the parameter $A$ reduces the motor velocity through
  $$
  V_+(n_+) = V_{+,0}e^{-A n_+}, \qquad V_-(n_-) = -V_{+,0}e^{-A n_-}.
  $$
  This produces stronger accumulation (larger peaks) and reduced transport.

- **Jamming effect:** the peak values increase with $A$:
  - $A=0$: $\max_x n_+ = 0.040921$, $\max_x n_- = 0.059679$
  - $A=2$: $\max_x n_+ = 0.044001$, $\max_x n_- = 0.062182$
  - $A=7$: $\max_x n_+ = 0.056580$, $\max_x n_- = 0.073114$
  This indicates a higher risk of jamming for larger $A$.

- **Transport reduction:** both the flux at the center $J(x=10,t)$ and the spatially averaged flux
  $$
  \bar J(t)=\frac{1}{L}\int_0^L J(x,t)\,dx
  $$
  decrease as $A$ increases, meaning that density-dependent slowing reduces the overall transport.

### Simplifications in the model
- The model is 1D and uses constant parameters (except for the velocity change in Hypothesis I).
- Boundary conditions prescribe fixed inflow/loading at $x=0$ and $x=L$.
- The numerical method is first-order in space for advection (diffusive but robust).

### Possible further investigations
- Perform a finer parameter study (more values of $A$) and determine a threshold where jamming becomes significant.
- Use a refined numerical scheme (higher order or less diffusive flux) to resolve sharper jam fronts.
- Compare model predictions to experimental/literature data to estimate parameters and validate the jamming mechanism.
